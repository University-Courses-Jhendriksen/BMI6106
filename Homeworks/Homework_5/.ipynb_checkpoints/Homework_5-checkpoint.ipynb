```r
# How many levels do categorical variables have?
cat_var <- c("gender","hypertension","heart_disease","ever_married","work_type","Residence_type","smoking_status","stroke")

categorical_levels <- as.data.frame(matrix(nrow=8,ncol=1))
colnames(categorical_levels) <- c("Number of Levels")
rownames(categorical_levels) <- cat_var
for(k in cat_var){
    levels <- unique(stroke[,k])
    categorical_levels[k,1] <- length(levels)
    cat(sprintf("\n%s: %s",k,levels))
}
categorical_levels
```

```r
# there are errors/miscounts in the levels such as male vs Male and 10 in heart disease
male <- stroke$gender == "male"
stroke$gender[male] = "Male"
female <- stroke$gender == "female"
stroke$gender[female] = "Female"
weird10 = stroke$heart_disease == 10
stroke$heart_disease[weird10] = sample(0:1,1)

# repeat procedure from earlier to create dataframe with number of levels
cat_var <- c("gender","hypertension","heart_disease","ever_married","work_type","Residence_type","smoking_status","stroke")

categorical_levels <- as.data.frame(matrix(nrow=8,ncol=1))
colnames(categorical_levels) <- c("Number of Levels")
rownames(categorical_levels) <- cat_var
for(k in cat_var){
    levels <- unique(stroke[,k])
    categorical_levels[k,1] <- length(levels)
    cat(sprintf("\n%s: %s",k,levels))
}
categorical_levels[2,1] = 2
categorical_levels
```

```r
# Distribution of categorical data
library("plyr")
gender_count = count(stroke$gender)
gender_count <- gender_count %>% mutate(percentage = freq/length(stroke$gender))
colnames(gender_count) = c("gender","count","frequency")

hyper_count = count(stroke$hypertension)
hyper_count <- hyper_count %>% mutate(percentage = freq/length(stroke$gender))
colnames(hyper_count) = c("hypertension","count","frequency")

heart_count = count(stroke$heart_disease)
heart_count <- heart_count %>% mutate(percentage = freq/length(stroke$gender))
colnames(heart_count) = c("heart disease","count","frequency")

married_count = count(stroke$ever_married)
married_count <- married_count %>% mutate(percentage = freq/length(stroke$gender))
colnames(married_count) = c("ever married","count","frequency")

work_count = count(stroke$work_type)
work_count <- work_count %>% mutate(percentage = freq/length(stroke$gender))
colnames(work_count) = c("work type","count","frequency")

residence_count = count(stroke$Residence_type)
residence_count <- residence_count %>% mutate(percentage = freq/length(stroke$gender))
colnames(residence_count) = c("residence type","count","frequency")

smoking_count = count(stroke$smoking_status)
smoking_count <- smoking_count %>% mutate(percentage = freq/length(stroke$gender))
colnames(smoking_count) = c("smoking status","count","frequency")

stroke_count = count(stroke$stroke)
stroke_count <- stroke_count %>% mutate(percentage = freq/length(stroke$gender))
colnames(stroke_count) = c("stroke","count","frequency")

gender_count
hyper_count
heart_count
married_count
work_count
residence_count
smoking_count
stroke_count
```

```r
# CENTRAL TENDENCY STATISTICS
# - code of Bob
# FUNCTIONS
#
getmode <- function(v) {
   uniqv <- unique(v)
   uniqv[which.max(tabulate(match(v, uniqv)))]
}
gm_mean = function(x, na.rm=TRUE){
  exp(sum(log(x[x > 0]), na.rm=na.rm) / length(x))
}

# CENTRAL TENDENCY STATISTICS

# BMI
bmi = as.numeric(stroke$bmi[stroke$bmi !='N/A'])
mean(bmi)
median(bmi)
getmode(bmi)
gm_mean(bmi)
quantile(bmi,probs = c(0.25,0.75))

# AVG GLUCOSE LEVEL 
avg_glucose_level = as.numeric(stroke$avg_glucose_level)
mean(avg_glucose_level)
median(avg_glucose_level)
getmode(avg_glucose_level)
gm_mean(avg_glucose_level)
quantile(avg_glucose_level,probs = c(0.25,0.75))

# AGE
age = as.numeric(stroke$age)
mean(age)
median(age)
getmode(age)
gm_mean(age)
quantile(age,probs = c(0.25,0.75))
```
