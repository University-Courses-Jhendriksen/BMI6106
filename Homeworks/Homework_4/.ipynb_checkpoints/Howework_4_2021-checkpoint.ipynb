##### 2. X is a continuous random variable with mean μ = 40, Find and plot the density distribution (pdf) with three different standard deviations (2,6,12) (Please draw the curves on the same figure with different colors and a legend - hint: see function lines()) for:

a. P(x < 40)<br><br>
b. P(x > 21)

c. What conclusions can you draw from these plots regarding the effect of the SD on the distribution?

```r
##Code
# assuming normal distribution about mean
samp = seq(0,80,by = 0.01)
pmf40sd2 = dnorm(samp,40,2)
pmf40sd6 = dnorm(samp,40,6)
pmf40sd12 = dnorm(samp,40,12)
plot(samp,pmf40sd2,type='l',col='blue')
lines(samp,pmf40sd6,col='green')
lines(samp,pmf40sd12,col='red')
legend("topright",c("std dev = 2","std dev = 6","std dev = 12"),fill=c("blue","green","red"))
# a) Probability of x < 40
pl40sd2 = pnorm(40,40,2)
pl40sd6 = pnorm(40,40,6)
pl40sd12 = pnorm(40,40,12)
cat(sprintf("Probablity of less than 40 w/std dev = 2: %s\n",pl40sd2))
cat(sprintf("Probablity of less than 40 w/std dev = 6: %s\n",pl40sd6))
cat(sprintf("Probablity of less than 40 w/std dev = 12: %s\n\n",pl40sd12))
abline(v = 40)
# b) Probability of x > 21
pg21sd2 = 1-pnorm(21,40,2)
pg21sd6 = 1-pnorm(21,40,6)
pg21sd12 = 1-pnorm(21,40,12)
cat(sprintf("Probablity of greater than 21 w/std dev = 2: %s\n",pg21sd2))
cat(sprintf("Probablity of greater than 21 w/std dev = 6: %s\n",pg21sd6))
cat(sprintf("Probablity of greater than 21 w/std dev = 12: %s\n\n",pg21sd12))
abline(v=21,col="purple")
legend("topleft",c("X = 40","X = 21"),fill = c("black","purple"))
```

Part C - Conclusions

From the plots it is clear that the standard deviation affects the spread of the data and how likely it is to have a result a certain distance away from the mean. This is evident from the calculated probabilites - even though the probability of any number at the mean or lower is 0.5 for all distributions, when looking at the probability of X > 21 you can see that the spread of the data is different in that at a standard deviation of 2 and 6 basically 100% of the time it will be greater than 21 but this will only happen 94% of the time when the standard deviation is 12.

##### 3. There are 6 cars in a car shop out which 3 are defective. If 2 cars are picked randomly, 

#### <font color=blue>Find the probability that at least one is defective.</font>

1. Use the density function
2. Use the R code to compare the results

```r
##Code

# this is a hypergeometric distribution because we are sampling the cars without replacing them
# k = num successes, n =  num trials, N = population size, K = successful objects
k3 = 1:2
n3 = 2
N3 = 6
K3 = 3
pHyp1or2 <- (choose(K3,k3)*choose(N3-K3,n3-k3))/choose(N3,n3)
pHypAtLeast2 = sum(pHyp1or2)
cat(sprintf("Density function probability of at least 1 defective = %s\n",pHypAtLeast2))
# by function in R
pHypR <- sum(dhyper(k3,K3,N3-K3,n3))
cat(sprintf("\nProbability of at least 1 defective using R code = %s",pHypR))
```

##### 4. In the past, for every attempt to make a call there was a 70% probability of getting the call. 

#### <br><font color=blue> a. Calculate the probability of having  12 successes in 20 attempts. 
1. Use the density function<br>
2. Use the R code to compare the results<br> 
    
#### <br><font color=blue>b. Plot the distribution and describe the shape</font>

```r
# binomial distribution - get the call or don't
p4 = 0.7
q4 = 1-p4
prob12in20 = choose(20,12)*(p4^12)*(q4^(20-12))
prob12in20R = dbinom(12,20,prob=0.7)
cat(sprintf("Probability of 12 calls out of 20 tries by distribution is: %s and by the R code is: %s",prob12in20,prob12in20R))
# plotting distribution
binomDist20 = dbinom(0:20,20,0.7)
plot(0:20,binomDist20,ylab="probability",xlab="number of successes",col="blue",type="l")
abline(v=12)
legend("topleft",c("distribution","12 successes"),fill=c("blue","black"))
mean4b = 20*p4
stddev4b = sqrt(20*p4*q4)
cat(sprintf("\n\nThis is a binomial distribution with an average of %s and a standard deviation %s",mean4b,stddev4b))
```

# Making use of a beta distribution

There is a bag containing an unknown number of white balls and red balls:

After drawing and replacing a ball 10 times you have drawn a red 7 times and a white 3 times.

What is the probability of drawing a red ball?

You draw 3 more balls - Red, Red, White,White - how does this change the distribution and probability?

Plot the distribution

```r
# using distribution
red0 = 7
white0 = 3
distP = red0/(red0+white0)
cat(sprintf("Most likely probability of drawing a red based of initial draws = %s",distP))
redNew = 2
whiteNew = 2
distUp = (red0+redNew)/(red0+redNew+white0+whiteNew)
cat(sprintf("\n\nMost likely probability of drawing a red after inclusion of second set of draws = %s",distUp))
# using built in functions
se = seq(0,1,by = 0.01)
betaP = dbeta(se,red0,white0)
betaUp = dbeta(se,red0+redNew,white0+whiteNew)
plot(se,betaUp,type = 'l',col = 'red',main = "Beta distribution for probability of drawing red ball",ylab = "probability density",xlab = "probability")
lines(se,betaP,type = 'l')
legend("topleft",c("After Initial draws","Including 2nd set of draws"),fill = c("black","red"))
cat(sprintf("\n\nThe inclusion of more draws both shifts the peak of the distribution towards it's true probabilty but also narrows\nit as more observations have been made"))
```
