### 1 From the heart disease dataset heart.csv calculate mean, SD, quantiles, min and max for each continuous variable, the result should be stored in a named list.

```r
heart <- read.csv(file = 'heart.csv',header = T)
head(heart)

age <- heart[,1]
trestbps <- heart[,4]
chol <- heart[,5]
thalach <- heart[,8]
oldpeak <- heart[,10]

means <- list(mean(age),mean(trestbps),mean(thalach),mean(oldpeak))
SDs <- list(sd(age),sd(trestbps),sd(thalach),sd(oldpeak))
quantiles <- list(quantile(age),quantile(trestbps),quantile(thalach),quantile(oldpeak))
mins <- list(min(age),min(trestbps),min(thalach),min(oldpeak))
maxes <- list(max(age),max(trestbps),max(thalach),max(oldpeak))

meansV <- c(mean(age),mean(trestbps),mean(thalach),mean(oldpeak))
SDsV <- c(sd(age),sd(trestbps),sd(thalach),sd(oldpeak))
quantilesV <- c(quantile(age),quantile(trestbps),quantile(thalach),quantile(oldpeak))
minsV <- c(min(age),min(trestbps),min(thalach),min(oldpeak))
maxesV <- c(max(age),max(trestbps),max(thalach),max(oldpeak))

results <- list()
results[["Mean"]] <- meansV
results[["SD"]] <- SDsV
results[["Quantiles"]] <- quantilesV
results[["Min"]] <- minsV
results[["Max"]] <- maxesV

results
```

### 2 Calculate the same statistical summaries in list form as above but by split by sex levels (male/female)

```r
M = heart$sex == 1
F = heart$sex != 1
heartM <- heart[M,]
heartF <- heart[F,]


ageM <- heartM[,1]
trestbpsM <- heartM[,4]
cholM <- heartM[,5]
thalachM <- heartM[,8]
oldpeakM <- heartM[,10]

ageF <- heartF[,1]
trestbpsF <- heartF[,4]
cholF <- heartF[,5]
thalachF <- heartF[,8]
oldpeakF <- heartF[,10]


meansM <- c(mean(ageM),mean(trestbpsM),mean(thalachM),mean(oldpeakM))
SDsM <- c(sd(ageM),sd(trestbpsM),sd(thalachM),sd(oldpeakM))
quantilesM <- c(quantile(ageM),quantile(trestbpsM),quantile(thalachM),quantile(oldpeakM))
minsM <- c(min(ageM),min(trestbpsM),min(thalachM),min(oldpeakM))
maxesM <- c(max(ageM),max(trestbpsM),max(thalachM),max(oldpeakM))

meansF <- c(mean(ageF),mean(trestbpsF),mean(thalachF),mean(oldpeakF))
SDsF <- c(sd(ageF),sd(trestbpsF),sd(thalachF),sd(oldpeakF))
quantilesF <- c(quantile(ageF),quantile(trestbpsF),quantile(thalachF),quantile(oldpeakF))
minsF <- c(min(ageF),min(trestbpsF),min(thalachF),min(oldpeakF))
maxesF <- c(max(ageF),max(trestbpsF),max(thalachF),max(oldpeakF))


resultsMale <- list()
resultsMale[["Mean"]] <- meansM
resultsMale[["SD"]] <- SDsM
resultsMale[["Quantiles"]] <- quantilesM
resultsMale[["Min"]] <- minsM
resultsMale[["Max"]] <- maxesM

resultsFemale <- list()
resultsFemale[["Mean"]] <- meansF
resultsFemale[["SD"]] <- SDsF
resultsFemale[["Quantiles"]] <- quantilesF
resultsFemale[["Min"]] <- minsF
resultsFemale[["Max"]] <- maxesF

print("Male")
resultsMale
print("Female")
resultsFemale
```

### 6 Create a function that calculates the index of a character vector and its sum (e.g. abc = 1,2,3 and 6)

```r
charSum = function(char){
    charList <- letters
    char <- strsplit(char,split="")
    char = unlist(char)
    indChar = 0
    for(x in 1:length(char)){
        k = 0
        for(k in 1:length(charList)){
            match <- charList[k] == char[x]
            if(match == T){
                ind = k
            }
            }
        indChar[x] <- ind
        }
    charSum <- list()
    charSum[["index"]] <- indChar
    charSum[["sum"]] <- sum(indChar)
    return(charSum)
}

charSum("abc")
charSum("jordan")
```
