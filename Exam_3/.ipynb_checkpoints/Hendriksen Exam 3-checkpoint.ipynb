```r
library(tidyverse) 
library(gridExtra) 
library(mice)
library(ggplot2)
options(repr.plot.width = 12, repr.plot.height=8)

# prcomp and scree plots
pca_cols = subset(scaled_county_data, select = -c(log_paamv) )
county_PCA = prcomp(pca_cols)
par(mfrow=c(1,2))
screeplot(county_PCA,type = "lines")
summary(county_PCA)

# direct computation
counties.cov <- cov(scaled_county_data)
counties.eigen <- eigen(counties.cov)
str(counties.eigen)
plot(counties.eigen$values)
lines(c(9,9),c(0,14))

phi <- counties.eigen$vectors[,1:5]
phi <- -phi
row.names(phi) <- names(scaled_county_data)
colnames(phi) <- c("PC1", "PC2","PC3","PC4","PC5")

PC1 <- as.matrix(scaled_county_data) %*% phi[,1]
PC2 <- as.matrix(scaled_county_data) %*% phi[,2]
PC3 <- as.matrix(scaled_county_data) %*% phi[,3]
PC4 <- as.matrix(scaled_county_data) %*% phi[,4]
PC5 <- as.matrix(scaled_county_data) %*% phi[,5]

# Create data frame with Principal Components scores
PC <- data.frame(PC1, PC2,PC3,PC4,PC5)

# Plot Principal Components for each State
options(repr.plot.width = 16, repr.plot.height=8)
layout(matrix(1:2, ncol=2))
plot1 = ggplot(PC, aes(PC1, PC2)) + 
  modelr::geom_ref_line(h = 0) +
  modelr::geom_ref_line(v = 0) +
  geom_point(size = 3) +
  xlab("First Principal Component") + 
  ylab("Second Principal Component") 
plot2 = ggplot(PC, aes(PC1, PC3)) + 
  modelr::geom_ref_line(h = 0) +
  modelr::geom_ref_line(v = 0) +
  geom_point(size = 3) +
  xlab("First Principal Component") + 
  ylab("Third Principal Component")
plot3 = ggplot(PC, aes(PC1, PC4)) + 
  modelr::geom_ref_line(h = 0) +
  modelr::geom_ref_line(v = 0) +
  geom_point(size = 3) +
  xlab("First Principal Component") + 
  ylab("Fourth Principal Component")
plot4 = ggplot(PC, aes(PC2, PC3)) + 
  modelr::geom_ref_line(h = 0) +
  modelr::geom_ref_line(v = 0) +
  geom_point(size = 3) +
  xlab("Second Principal Component") + 
  ylab("Third Principal Component")
plot5 = ggplot(PC, aes(PC2, PC4)) + 
  modelr::geom_ref_line(h = 0) +
  modelr::geom_ref_line(v = 0) +
  geom_point(size = 3) +
  xlab("Second Principal Component") + 
  ylab("Fourth Principal Component")
plot6 = ggplot(PC, aes(PC2, PC5)) + 
  modelr::geom_ref_line(h = 0) +
  modelr::geom_ref_line(v = 0) +
  geom_point(size = 3) +
  xlab("Second Principal Component") + 
  ylab("Fifth Principal Component")
grid.arrange(plot1,plot2,plot3,plot4,plot5,plot6,ncol=3)
```

```r
# Biplots for main components
options(repr.plot.width = 16, repr.plot.height=16)
biplot(county_PCA,expand = 1.5,xlim = c(-0.15,0.15))
```

The results of PCA reveal that almost 30% of the variance in log-PAAMV is attributable to a single direction along the first principal component. By the 10th of 46 principal components over 73% of variance is accounted for and the variance for each component has dropped to ~ =< 2%. This is evident in the scree plot which shows a final small elbow at PC9. The biplot shows that a large number of the explanatory variables are oriented in relatively similar direction to PC1. No distinct clusters form when looking at different combinations of PC1-PC5 as PCA plots though counties that are distinct from the average can be seen. 

```r
# Linear regression using the 9 variables most aligned with the first 9 PCA components
var_names = rownames(county_PCA$rotation)
best_var = c()
best_var_cor = list()
for(k in 1:9){
    col_rot = abs(county_PCA$rotation[,k])
    best_ind = which(abs(county_PCA$rotation[,k]) == max(col_rot), arr.ind = TRUE)
    best_ind2 = which(county_PCA$rotation[,k] == max(county_PCA$rotation[,k]), arr.ind = TRUE)
    best_var[k] = var_names[best_ind]
    best_var_cor[k] = county_PCA$rotation[best_ind,k]
}
best_var
best_var_cor
pca_lin_data = scaled_county_data[best_var]
pca_lin_data$log_paamv = scaled_county_data$log_paamv
pca_lin_reg = lm(log_paamv ~ ., data = pca_lin_data)
summary(pca_lin_reg)

pca_trans_data = data.frame(county_PCA$x[,1:9])
pca_trans_data$log_paamv = scaled_county_data$log_paamv
pca_trans_lin_reg = lm(log_paamv~.,data=pca_trans_data)
summary(pca_trans_lin_reg)

model_comp = AIC(lin_reg,pca_lin_reg,pca_trans_lin_reg)
model_comp
```

Because of the final elbow at PC9 and small amount of variance explained by the PC following PC9 focus was placed on using data from PC1-PC9 for linear regression. Using the projections of the explanatory variables onto the eigenvectors of the principal components we isolated the variable corresponding most to each of PC1-PC9 and used those for linear regression. This resulted in a model with an R^2 value of 0.75 and an AIC of 1218 versus 0.899 and 500 when using all of the explanatory variables. While 0.75 was still respectable this model was substantially worse than the original.

Next, a model was built using the data transformed by the eigenvectors, or PC1-PC9 directly. This model had 7 signficant explanatory variables out of 9 and an R^2 value of 0.883 with an AIC of 591. Compared to the original model using 46 explantory variables this captures a signficant amount of the data and explanation that the original does with a much simpler model that is likely to be more generalized and perform better on new data.

$2.$ We will consider data from Stanford Heart Transplant Program, reported by Crowley and Hu (1977). The research question was whether receiving a heart transplant affected survival. 
Starting in November of 1967, 103 patients in need of heart transplant were accepted in the program. Sixty-nine of these patients received a transplant. Seventy-five people had an event, while 28 subjects were censored.  For each person in the program we have the date and age at the acceptance to the program, their de-identified ID, transplant status (and time of transplant if transplant was received), if a subject had a prior bypass surgery, survival time, and if the event occurred or if the subject was censored. These data are included as part of the R survival package (dataset named heart). 

Transplant status is an example of a time-varying covariate. Everybody starts as pre-transplant, then some people live to receive a transplant and others die prior to being able to receive one. If two people died at the same time, but one received a transplant 1 year prior to his death, and the other received a transplant 1 day prior to death, we cannot treat them in the same way. We must appropriately attribute a person’s time to either the “pre-transplant” or “post-transplant” period.
The way to deal with this data is to have 2 records for the patients in the study. One for the pre-transplant time, and another one for post-transplant (those subjects who did not get a transplant will only have one record).
